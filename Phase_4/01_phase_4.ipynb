```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.random.rand(3, 3)
array3 = np.zeros((4, 4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array1)
np.save('array3.npy', array1)
```

```python
loaded_array1 = np.load('array1.npy')
print("Array 1 : \n", loaded_array1)

loaded_array2 = np.load('array2.npy')
print("Array 2 : \n", loaded_array2)

loaded_array3 = np.load('array3.npy')
print("Array 3 : \n", loaded_array3)
```

```python
try:
    logo = np.load('numpy_logo.npy')

    # Display
    plt.figure(figsize=(8, 6))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    dark_logo = 1 - logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Dark Numpy Logo")
    plt.grid(False)

except FileNotFoundError:
    print("Numpy logo file not found")
```
